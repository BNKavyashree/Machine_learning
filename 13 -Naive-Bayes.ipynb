what is the probability of getting queen? total there will be 52 cards in deck among which 4 are queens so,p(queen)= 4/52 = 1/13.
pick a random card, you know it is a diamond. Now whats the probability of that card being a queen?
this is conditional probability and represented by P(A/B) - Probability of event A knowing that event B has occured.
total diamonds = 13
queen =1 
p(queen/diamond) = 1/13

Bayes equation:
P(A/B) = P(B/A)*P(A)/P(B)

p(queen/diamond)= p(diamond/Queen)*p(Queen)/P(diamond) = 1/4 * 1/13/ 1/4 = 1/13

In below code we will be finding the probability of survival: P(Survived/Male&Class&Age&Cabin&Fare)

The reason this algorithm is called naive because, We make a naive assumption that features such as male, class, age, cabin, fare etc. are independent of each other.

```python
import pandas as pd
df = pd.read_csv(r'Data\titanic.csv')
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp','Parch','Ticket', 'Cabin','Embarked'], axis='columns', inplace=True)
df.head()
```

```python
target =df.Survived
input = df.drop('Survived', axis='columns')
```

```python
dummies = pd.get_dummies(input.Sex,dtype='int')
dummies.head()
```

```python
input = pd.concat([input, dummies], axis='columns')
input.head()
```

```python
input.drop('Sex',axis ='columns', inplace=True)
input.head()
```

```python
input.columns[input.isna().any()] #check if there are any na values in the columns
```

```python
input.Age[:10]
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.2)
len(input),len(X_train), len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB #guassianNb is used when we have the data in continuous form
model = GaussianNB()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```
